```python
def stringquery(ilocdata):
    string = ""
    #get the columns that i want
    cols = ['banner_pos','hour','site_id','site_domain','site_category','app_id','app_domain','app_category']
    for t,i in enumerate(cols):
        if t == (len(cols)-1):
            string+= ' {}==\'{}\''.format(i,ilocdata[i])
        else:
            string+= ' {}==\'{}\' &'.format(i,ilocdata[i])
    return string
stringquery(X7to10pm.iloc[-1])
```
